```python
import os
import pandas as pd

# Create an empty DataFrame to hold all data
all_months_data = pd.DataFrame()

# Directory containing the CSV files
folder_path = 'D:/python/JULY_COHORT-master/lesson4b(Sales Analysis)/Sales Analysis/panda_practical/Sales_Data'

# List all CSV files in the directory
files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Loop through each file and read its contents
for file in files:
    # Construct the file path correctly
    file_path = os.path.join(folder_path, file)
    try:
        print(f"Processing file: {file_path}")
        df = pd.read_csv(file_path)
        all_months_data = pd.concat([all_months_data, df], ignore_index=True)
    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found.")
    except pd.errors.EmptyDataError:
        print(f"Warning: The file {file_path} is empty.")
    except pd.errors.ParserError:
        print(f"Error: The file {file_path} could not be parsed.")
    except Exception as e:
        print(f"An error occurred while reading {file_path}: {e}")

# Display the first few rows of the combined DataFrame
print(all_months_data.head(12))
```

```python
import matplotlib.pyplot as plt

# Ensure 'Hour' column exists. If not, create it from a timestamp column if available
if 'Hour' not in all_data.columns:
    if 'Order Date' in all_data.columns:
        # Assuming 'Order Date' is a datetime column
        all_data['Hour'] = pd.to_datetime(all_data['Order Date']).dt.hour

# Ensure 'Count' column exists. If not, create it
if 'Count' not in all_data.columns:
    all_data['Count'] = 1  # Assuming each row represents one occurrence

# Group by 'Hour' and count occurrences
hourly_counts = all_data.groupby('Hour').count()['Count']

# Extract keys (hours) and values (counts)
keys = hourly_counts.index
counts = hourly_counts.values

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(keys, counts, marker='o')
plt.xticks(keys)
plt.xlabel('Hour of the Day')
plt.ylabel('Count of Orders')
plt.title('Number of Orders per Hour')
plt.grid(True)
plt.show()
```
